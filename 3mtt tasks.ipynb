```python
##Define a function that takes a list of sales amounts. Use a for loop with an if statement to print amounts over 100.

def purchases_100(salesAmounts):
    count = 0
    for salesAmount in salesAmounts:
        if salesAmount > 100:
            count += 1
    print (count)

purchases_100([50, 150, 200, 10])
```
